```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Read Inside Airbnb data").getOrCreate()
```

```python
listings = spark.read.csv("/Volumes/bootcamp_students/zachy_w_brett_coleman/airbnbdata/listings.csv.gz",
        header=True,
        inferSchema=True,
        sep=",",
        quote='"',
        escape='"',
        multiLine=True,
        mode="PERMISSIVE"
)
```

```python
# Look at all of the fields in the dataset
for field in listings.schema:
    print(field)
```

```python
neighbourhoods = listings.select(listings.neighbourhood_cleansed)
neighbourhoods.show(20, truncate=False)
```
